```python
import pandas as pd
import numpy as np
```

```python
rotten_movie_data=pd.read_csv('/content/Python_Project_File/rotten_tomatoes_movies.csv')
rotten_movie_data.info()
```

```python
rotten_movie_data.drop(['rotten_tomatoes_link','movie_info','critics_consensus','content_rating',\
                        'rotten_tomatoes_link','tomatometer_rotten_critics_count','tomatometer_fresh_critics_count',\
                        'tomatometer_top_critics_count','audience_count','audience_rating','audience_status',\
                        'tomatometer_count','tomatometer_status','production_company','runtime','streaming_release_date',\
                        'actors','authors','directors','genres'],axis=1,inplace=True)
```

```python
rotten_movie_clean_miss=rotten_movie_data.loc[rotten_movie_data.original_release_date.notnull()].copy()
rotten_movie_clean_miss['original_release_date']=pd.to_datetime(rotten_movie_clean_miss['original_release_date'])
```

```python
def find_year(date):
    return date.year
```

```python
rotten_movie_clean_miss['year']=rotten_movie_clean_miss.apply(lambda row:find_year(row['original_release_date']),axis=1)
rotten_movie_clean_miss.drop(['original_release_date'],axis=1,inplace=True)
rotten_movie_clean_miss['year']=rotten_movie_clean_miss['year'].astype(int)
```

```python
rotten_movie_clean_miss.head()
```

```python
movie_imdb=pd.read_csv('/content/Python_Project_File/IMDb movies.csv')
```

```python
movie_imdb.drop(['title','reviews_from_users','reviews_from_critics','director',\
                 'actors','writer','production_company','description'],axis=1,inplace=True)
movie_imdb.rename({'worlwide_gross_income':'worldwide_gross_income','avg_vote':'imdb_score'},axis=1,inplace=True)
#movie_imdb.head(2)
```

```python
movie_imdb.columns
```

```python
movie_imdb_clean=movie_imdb.loc[movie_imdb.worldwide_gross_income.notnull()].copy()
```

```python
movie_imdb_clean['year']=movie_imdb_clean['year'].astype(int)
```

```python
movie_imdb_clean_income=movie_imdb_clean.loc[movie_imdb_clean.worldwide_gross_income.str.startswith('$')].copy()
```

```python
def remove_doller(row):
    return row.replace('$','')
```

```python
movie_imdb_clean_income['worldwide_gross_income_num']=movie_imdb_clean_income.apply(lambda row:remove_doller(row['worldwide_gross_income']),axis=1)
movie_imdb_clean_income.drop(['worldwide_gross_income'],axis=1,inplace=True)
movie_imdb_clean_income.rename({'worldwide_gross_income_num':'worldwide_gross_income','original_title':'movie_title'},axis=1,inplace=True)
movie_imdb_clean_income['worldwide_gross_income']=movie_imdb_clean_income['worldwide_gross_income'].astype(float)
movie_imdb_clean_income['worldwide_gross_income']=movie_imdb_clean_income['worldwide_gross_income']/1000000
```

```python
def movie_age(year):
    return 2021-year
```

```python
movie_imdb_clean_income['movie_age']=movie_imdb_clean_income.apply(lambda x:movie_age(x['year']),axis=1)
```

```python
movie_imdb_clean_income.head(1)
#rotten_movie_clean_miss.head(1)
```

```python
df=pd.merge(movie_imdb_clean_income,rotten_movie_clean_miss,on='movie_title',how='inner')
```

```python
movie_imdb_clean_income.loc[movie_imdb_clean_income.movie_title=='Cleopatra']
```

```python
rotten_movie_clean_miss.loc[rotten_movie_clean_miss.movie_title=='Cleopatra']
```

```python
df.loc[df.movie_title=='Cleopatra']
```

```python
df_final=pd.merge(movie_imdb_clean_income,rotten_movie_clean_miss,on=['movie_title','year'],how='inner')
```

```python
df_final.drop(['imdb_title_id','date_published','votes','budget','usa_gross_income','metascore'],axis=1,inplace=True)
```

```python
df_final.head(5)
```

```python
df_final['imdb_score']=df_final.apply(lambda x:x['imdb_score']*10,axis=1)
```

```python
df_final
```

```python
df_final.to_csv('/content/Python_Project_File/combined_final_data.csv')
```

In this part of the assignment, you are supposed to take the raw imdb movie dataset.

Clean the worldwide gross income column the way we had cleaned it in the lesson on Data Cleaning and Manipulation.

Save the dataframe in the file `imdb_income_budget.csv`.

```python
imdb_movie=pd.read_csv('/content/Python_Project_File/IMDb movies.csv')
```

```python
imdb_movie.columns
```

```python
imdb_movie_clean_nan=imdb_movie.loc[imdb_movie['worlwide_gross_income'].notnull()].copy()
imdb_movie_clean_income=imdb_movie_clean_nan.loc[imdb_movie_clean_nan.worlwide_gross_income.str.startswith('$')].copy()
```

```python
def remove_doller(raw):
    return raw.replace('$','')
```

```python
imdb_movie_clean_income['worldwide_gross_income']=imdb_movie_clean_income.apply(lambda x:remove_doller(x['worlwide_gross_income']),axis=1)
imdb_movie_clean_income.drop(['worlwide_gross_income'],axis=1,inplace=True)
imdb_movie_clean_income['worldwide_gross_income']=imdb_movie_clean_income['worldwide_gross_income'].astype(float)
```

```python
imdb_movie_budget_clean=imdb_movie_clean_income.loc[imdb_movie_clean_income['budget'].notnull()].copy()
imdb_movie_budget_clean_data=imdb_movie_budget_clean.loc[imdb_movie_budget_clean.budget.str.startswith('$')].copy()
```

```python
imdb_movie_budget_clean_data['budget']=imdb_movie_budget_clean_data.apply(lambda x:remove_doller(x['budget']),axis=1)
imdb_movie_budget_clean_data['budget']=imdb_movie_budget_clean_data['budget'].astype(float)
```
